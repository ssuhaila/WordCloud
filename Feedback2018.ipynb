## Steps involved:-

1. Import the neccessary package
2. Read the feedbacks from a .txt file
3. Remove any characters (some of them used * or - as bulletpoints)
4. Finding the WordCount in the feedback
5. Generate the WordCloud
6. Find the most frequent words used in this feedbacks (Top 10)

### 1. Import the packages needed

```python
from wordcloud import WordCloud, STOPWORDS, ImageColorGenerator
import matplotlib.pyplot as plt 
import pandas as pd 
import re
```

### 2. Read the .txfile

```python
f = open("feedback2018.txt", "r")
feedback_ori = f.read()
print(feedback_ori)
```

### 3. Remove any characters

```python
#remove characters and numbers 
#Code from: https://stackoverflow.com/questions/44315941/regex-to-strip-all-numbers-and-special-characters-but-space-and-letters

clean_text=re.sub(r'([^a-zA-Z\s]+?)', '',  feedback_ori)
print (clean_text)
```

### 3. Finding the WordCount in the feedback

```python
#https://www.sanfoundry.com/python-program-count-number-words-characters-file/

num_words=0

for line in clean_text:
    words = line.split()
    num_words += len(words)
print("Number of words in the feedback:")
print(num_words)
```

### 4. Create the WordCloud

```python
#convert the stopwords into a dict 
stopwords = set(STOPWORDS)
```

```python
#adding my own name in the stopwords set
stopwords.add('Siti')
```

```python
#generate the wordcloud

wordcloud=WordCloud(background_color='white', stopwords = stopwords).generate(clean_text)
```

```python
#plotting the wordcloud

plt.figure(figsize=(10,10))
plt.imshow(wordcloud, interpolation='bilinear')

plt.axis("off")

plt.show()
```

### 5. Finding the most-frequent used words in the feedbacks

```python
### Finding the most frequent-used words

words = clean_text.split()

word_counter = {}

for word in words:
    if word not in stopwords:
        if word not in word_counter:
            word_counter[word] = 1
        else:
            word_counter[word] += 1

popular_words = sorted(word_counter, key = word_counter.get, reverse = True)

print (popular_words[:10])
```
